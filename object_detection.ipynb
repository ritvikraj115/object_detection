```python
from ultralytics import YOLO
import numpy as np
import cv2
```

```python
model = YOLO('yolov8n.pt')
```

```python
cap = cv2.VideoCapture(0)

# Check if the webcam is opened successfully
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

while True:
    # Read a frame from the webcam
    ret, image = cap.read()

    results = model(image)
    for r in results:
        l=[]
        int_list = list(map(int, r.boxes.cls))
        for i in int_list:
            l.append((r.names)[i])
            
        xywh=np.array(r.boxes.xywh)
        # Convert xywh format to xyxy format (x_min, y_min, x_max, y_max)
        xyxy = np.zeros_like(xywh)
        xyxy[:, 0] = xywh[:, 0] - xywh[:, 2] / 2  # x_min
        xyxy[:, 1] = xywh[:, 1] - xywh[:, 3] / 2  # y_min
        xyxy[:, 2] = xywh[:, 0] + xywh[:, 2] / 2  # x_max
        xyxy[:, 3] = xywh[:, 1] + xywh[:, 3] / 2  # y_max

    # Draw bounding boxes on the image
    for box in xyxy:
        x_min, y_min, x_max, y_max = map(int, box)
        color = (0, 255, 0)  # Green color
        thickness = 2
        cv2.rectangle(image, (x_min, y_min), (x_max, y_max), color, thickness)
    for i in range(len(xywh)):
        text =l[i]
        print(text)
        font = cv2.FONT_HERSHEY_SIMPLEX
        text_pos=(int(xywh[i][0]),int(xywh[i][1]))
        print(text_pos)
        font_scale = 0.3
        font_color = (255, 255, 255)  # White color in BGR
        font_thickness = 1
    
    # Add the text to the image
        cv2.putText(image, text, text_pos, font, font_scale, font_color, font_thickness)
    
    # Display the result
    cv2.imshow("Image", image)

    # Break the loop if 'q' key is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


# Release the webcam and close the window
cv2.destroyAllWindows()
```
